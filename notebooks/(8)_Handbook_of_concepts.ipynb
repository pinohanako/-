# Справочник понятий

**Дефолт** – невыполнение обязательств по договору займа. Обычно дефолтом считается неоплата по договору в течение 90 дней.


---

**PD** (probability of default) – вероятность дефолта.

---


**ПДН** или **показатель долговой нагрузки** (Debt-to-Income Ratio) - отношение ежемесячных платежей по долгам к общему ежемесячному доходу заемщика. Помогает оценить способность заемщика справляться с финансовыми обязательствами и является важным фактором при принятии решений о выдаче кредита или займа. Рассчитывается как процентное отношение суммы ежемесячных платежей по всем обязательствам заемщика (включая ипотеку, кредитные карты, автокредиты и другие долги) к его ежемесячному доходу.
Банки и кредитные организации часто используют DTI для оценки риска невыплаты кредита. Обычно считается, что чем ниже DTI, тем лучше, так как это указывает на то, что у заемщика есть достаточный доход для покрытия его обязательств.


---

**Revolving utilization** или **использование кредитного лимита**. Это показатель, который отражает процент использования кредитного лимита, доступного заемщику на его кредитной карте или другом кредитном счете с возобновляемым лимитом. Рассчитывается как отношение суммы, потраченной заемщиком на кредитной карте или счете, к общему доступному кредитному лимиту. Он показывает, какую часть доступного кредитного лимита заемщик использует на текущий момент.

---

**grade** (класс, категория) — это оценка, которая используется для классификации кредитоспособности заемщика или качества кредитного портфеля. Эта оценка помогает кредиторам и финансовым учреждениям быстро определить уровень риска, связанного с предоставлением кредита или займа.

  A: Это высший класс или категория. Заемщики с рейтингом "A" считаются наиболее надежными и имеют отличную кредитную историю. Они, как правило, имеют высокий доход, стабильную работу и низкий уровень риска невыплаты кредита.

  B: Заемщики с рейтингом "B" считаются хорошими заемщиками, но имеют некоторый уровень риска. Они могут иметь менее идеальную кредитную историю, более низкий доход или другие факторы, которые могут повлиять на их способность выплачивать кредит.

  C: Рейтинг "C" указывает на более высокий уровень риска. Заемщики с этим рейтингом могут иметь проблемы с кредитной историей, нестабильный доход или другие факторы, которые делают их менее привлекательными для кредиторов.

Эта система градации помогает кредиторам быстро оценивать кредитоспособность заемщиков и принимать решения о предоставлении кредитов. Она также используется для определения процентных ставок и условий кредита. Заемщики с более высоким рейтингом, как правило, получают более выгодные условия, в то время как заемщики с более низким рейтингом могут сталкиваться с более высокими процентными ставками и более строгими требованиями.

---
Статус ферификации **verification_status**:

Not Verified (не проверено): Это означает, что информация, предоставленная заемщиком, не была проверена кредитным учреждением. Это может быть связано с тем, что заемщик только что подал заявку и процесс проверки еще не завершен, или же заемщик не предоставил достаточную информацию для проверки.
Source Verified (проверено источником): Статус "Source Verified" указывает на то, что информация, предоставленная заемщиком, была проверена и подтверждена независимым источником, таким как работодатель, банк или другая финансовая организация. Это повышает достоверность информации и снижает риск мошенничества.
Verified (проверено): Этот статус означает, что вся информация, предоставленная заемщиком, была тщательно проверена кредитным учреждением. Это может включать в себя проверку дохода, кредитной истории, занятости и других факторов. Полное подтверждение информации повышает уверенность кредитора в кредитоспособности заемщика.

**Public record**

В Америке с 80-х годов ведется централизованный электронный учет всех произошедших сделок; называется это Public Records. Каждая единица недвижимости пронумерована и имеет свой PIN. Например, чтобы посмотреть все операции по определенной недвижимости (кто является владельцем недвижимости, есть ли на недвижимости долги, судебные иски и прочее) достаточно зайти на сайт public records. Так и вводите в Гугл public records и штаты или county например "Illinois cook county" (у каждого county есть свой клерк и сайт). Через пару недель после сделки вы должны увидеть свои имя и фамилию в качестве владельца. Если они будут неправильно записаны свяжитесь со своим юристом.

**Fully Paid Status Loan**: Кредит с полностью погашенным статусом.

**Charged Off Status Loan**: Кредит со статусом "списан".

Кредит со статусом "списан" означает, что кредитор или кредиторская организация списали долг как безнадежный и закрыли счет для будущих операций. Это негативный элемент, который появляется в кредитном отчете, когда кредитор перестает пытаться взыскать просроченный долг и признает его потерей.

Как правило, кредиторы списывают счета после шести пропущенных платежей, но это может произойти и раньше.

Когда счет становится просроченным: после пропущенного платежа ваш счет больше не считается действительным, а может быть помечен как "потенциально негативный" или подобным. В записи будет указан непогашенный остаток по счету и срок, в течение которого он не оплачивался, с указанием 30-дневных периодов до 180 дней.

Когда происходит списание: если счет остается просроченным в течение определенного периода, обычно 120-180 дней, кредитор может решить списать его как потерю. Статус и непогашенный остаток по-прежнему будут отображаться в кредитном отчете, но он будет отмечен как "списанный".
